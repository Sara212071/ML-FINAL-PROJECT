```python
StudentDF['Target'].replace('Dropout',0,inplace=True)
StudentDF['Target'].replace('Graduate',1,inplace=True)
StudentDF['Target'].replace('Enrolled',2,inplace=True)
StudentDF.head()
```

```python
X = StudentDF.drop(columns=['Target'])  # X contains all columns except the target column which is the y_label
Y = StudentDF['Target']  # y contains the target column (y_llabel)
```

```python
#checking the frequency of each class 
Freq=StudentDF['Target'].value_counts().plot(kind='bar',color=['red','blue','navy'],
                                          title='Students Success',figsize=(8, 6))

plt.xticks(range(3),['Dropout','Graduate','Enrolled'])
Freq.set_xlabel("Target",fontsize=16)
Freq.set_ylabel("Number of Students",fontsize=14)
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2,random_state=42)
```

```python
#use the random search to get the best hyperparamter for the random forest model

from sklearn.model_selection import RandomizedSearchCV

# Define the hyperparameter search space
param_dist2 = {
    'max_depth': [10, 20, 30, 40, None], 
    'n_estimators': [100, 200, 300, 400],
}

# Create a Random Forest classifier
forest = RandomForestClassifier(random_state=42)

# Perform random search with cross-validation
random_search2 = RandomizedSearchCV(forest, param_distributions=param_dist2, n_iter=10, cv=5, scoring='accuracy', random_state=42)
random_search2.fit(x_train, y_train)

# Get the best hyperparameters
best_max_depth2 = random_search2.best_params_['max_depth']
best_n_estimators2 = random_search2.best_params_['n_estimators']

# Train a model with the best hyperparameters
best_forest = RandomForestClassifier(max_depth=best_max_depth2, n_estimators=best_n_estimators2, random_state=42)
best_forest.fit(x_train, y_train)
y_pred = best_forest.predict(x_test)  


# Print the results
print("Best Max Depth:", best_max_depth2)
print("Best Number of Estimators:", best_n_estimators2)
print("Accuracy on training set before removing the outliers: {:.3f}".format(best_forest.score(x_train, y_train)))
print("Accuracy on test set before removing the outliers: {:.3f}".format(best_forest.score(x_test, y_test)))
```

```python
# generating the confusion matrix of the new decision tree model 
Confusion = confusion_matrix(y_test, y_pred)
CM=ConfusionMatrixDisplay(Confusion)
CM.plot()
plt.show()
```

```python
RFmodel = best_forest

# Create a learning curve
train_sizes, train_scores, test_scores = learning_curve(RFmodel, x_train, y_train, train_sizes=np.linspace(0.1, 1.0, 10), cv=5)

# Calculate mean and standard deviation of training and test scores
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)

# Plot the learning curve
plt.figure(figsize=(8, 6))
plt.title('Learning Curve ')
plt.xlabel('Training set size')
plt.ylabel('Score')


# Plot the mean training and test scores
plt.plot(train_sizes, train_scores_mean, color='r', label='Training Score')
plt.plot(train_sizes, test_scores_mean, color='g', label='Test Score')

# Fill the area between the mean and the standard deviation of the training and test scores
plt.fill_between(train_sizes, train_scores_mean - train_scores_std, train_scores_mean + train_scores_std, alpha=0.1, color='r')
plt.fill_between(train_sizes, test_scores_mean - test_scores_std, test_scores_mean + test_scores_std, alpha=0.1, color='g')


plt.legend()
plt.show()
```

```python
y_train_pred = RFmodel.predict(x_train)
y_test_pred = RFmodel.predict(x_test)

# Calculate training and test accuracy
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print(f"Training Accuracy: {train_accuracy:.2f}")
print(f"Test Accuracy: {test_accuracy:.2f}")

# Print the classification report for the training dataset
print("Classification Report for Training Data:")
print(classification_report(y_train, y_train_pred))

# Print the classification report for the test dataset
print("Classification Report for Test Data:")
print(classification_report(y_test, y_test_pred))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

# Define a list of potential k values to try
k_values = range(1, 20)

best_k = None
best_accuracy = 0.0

# Iterate through k values and find the best one
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    accuracy = cross_val_score(knn, x_train, y_train, cv=5).mean()
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_k = k

# Train the final model with the best k
best_knn = KNeighborsClassifier(n_neighbors=best_k)
best_knn.fit(x_train, y_train)
y_pred = best_knn.predict(x_test)

print("Best k:", best_k)
print("Accuracy on training set before removing the outliers: {:.3f}".format(best_knn.score(x_train, y_train)))
print("Accuracy on test set before removing the outliers: {:.3f}".format(best_knn.score(x_test, y_test)))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier

# Define a list of potential max_depth and min_samples_leaf values
max_depth_values = range(1, 20)
min_samples_leaf_values = range(1, 10)

best_max_depth = None
best_min_samples_leaf = None
best_accuracy = 0.0

# Iterate through parameter combinations and find the best one
for max_depth in max_depth_values:
    for min_samples_leaf in min_samples_leaf_values:
        dt = DecisionTreeClassifier(max_depth=max_depth, min_samples_leaf=min_samples_leaf)
        accuracy = cross_val_score(dt, x_train, y_train, cv=5).mean()
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_max_depth = max_depth
            best_min_samples_leaf = min_samples_leaf

# Train the final model with the best parameters
best_dt = DecisionTreeClassifier(max_depth=best_max_depth, min_samples_leaf=best_min_samples_leaf)
best_dt.fit(x_train, y_train)
y_pred = best_dt.predict(x_test)

print("Best max_depth:", best_max_depth)
print("Best min_samples_leaf:", best_min_samples_leaf)
print("Accuracy on training set before removing the outliers: {:.3f}".format(best_dt.score(x_train, y_train)))
print("Accuracy on test set before removing the outliers: {:.3f}".format(best_dt.score(x_test, y_test)))
```

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, roc_auc_score, auc

# Assuming y_test is a one-hot encoded matrix or a list of labels
n_classes = len(np.unique(y_test))

# Define the models list and their corresponding labels
models = [RFmodel, KNNmodel, DTmodel, Gmodel]
model_labels = ['Random Forest', 'Knn', 'Decision Tree', 'Gradient Boosting']

# Create a figure for the ROC curve comparison plot
plt.figure(figsize=(8, 6))

# Plot ROC curve for each model
for model, label in zip(models, model_labels):
    probas_ = model.predict_proba(x_test)  # Calculate class probabilities (predicted probabilities) for each class
    
    # Binarize the true labels for each class
    y_test_bin = label_binarize(y_test, classes=np.unique(y_test))
    
    # Initialize variables to store the aggregate ROC curve data
    mean_fpr = np.linspace(0, 1, 100)
    mean_tpr = 0
    
    # Compute the ROC curve data for each class
    for i in range(n_classes):
        fpr, tpr, _ = roc_curve(y_test_bin[:, i], probas_[:, i], drop_intermediate=True)
        mean_tpr += np.interp(mean_fpr, fpr, tpr)
    
    # Average the ROC curve data across classes
    mean_tpr /= n_classes
    
    # Calculate the AUC (Area Under the Curve) for the averaged ROC curve
    mean_auc = auc(mean_fpr, mean_tpr)
    
    # Plot the ROC curve for the current model with its AUC value as part of the label
    plt.plot(mean_fpr, mean_tpr, label=f'{label} AUC: {mean_auc:.3f}')

# Customize the plot
plt.legend(loc=4)  # Add a legend in the lower-right corner
plt.xlabel('False Positive Rate')  # Label the x-axis as False Positive Rate
plt.ylabel('True Positive Rate')  # Label the y-axis as True Positive Rate
plt.title('Multiclass ROC Curves for Model Comparison')
plt.show()  # Display the ROC curve comparison plot
```
