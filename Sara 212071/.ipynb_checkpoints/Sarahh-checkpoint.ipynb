```python
#i created an outlier detector using EllipticEnvelope
outlierDetector1 =  EllipticEnvelope(contamination=0.2)
#Fit the Elliptic Envelope model and predict outliers in the features X
result=outlierDetector1.fit_predict(X) 
```

```python
# i detected the outliers using another method which is the LocalOutlierFactor to compare
outlierDetector2 =  LocalOutlierFactor(n_neighbors=20)
result2=outlierDetector2.fit_predict(X)
```

```python
# i detected the outliers using a third method which is the IsolationForest to compare 
outlierDetector3 =  IsolationForest(n_estimators=100,random_state = 42)
result3=outlierDetector3.fit_predict(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2,random_state=42)
```

```python
#i used the random search to get the best hyperparamter for the random forest model

from sklearn.model_selection import RandomizedSearchCV

# Define the potential values for the number of trees in the forest and the maximum depth of the trees
param_dist2 = {
    'max_depth': [10, 20, 30, 40, None], 
    'n_estimators': [100, 200, 300, 400],
}

# i Created a Random Forest classifier
forest = RandomForestClassifier(random_state=42)

# i Used random search to find the best combination of hyperparameters
random_search2 = RandomizedSearchCV(forest, param_distributions=param_dist2, n_iter=10, cv=5, scoring='accuracy', random_state=42)
random_search2.fit(x_train, y_train)

# i got the best hyperparameters from the search and assigned them
best_max_depth2 = random_search2.best_params_['max_depth']
best_n_estimators2 = random_search2.best_params_['n_estimators']

# here i trained the model using the best hyperparameters
best_forest = RandomForestClassifier(max_depth=best_max_depth2, n_estimators=best_n_estimators2, random_state=42)
best_forest.fit(x_train, y_train)
y_pred = best_forest.predict(x_test)  


# i only printed the results
print("Best Max Depth:", best_max_depth2)
print("Best Number of Estimators:", best_n_estimators2)
print("Accuracy on training set before removing the outliers: {:.3f}".format(best_forest.score(x_train, y_train)))
print("Accuracy on test set before removing the outliers: {:.3f}".format(best_forest.score(x_test, y_test)))
```

```python
# this is the confusion matrix of the random forest model before removing the outliers
Confusion = confusion_matrix(y_test, y_pred)
CM=ConfusionMatrixDisplay(Confusion)
CM.plot()
plt.show()
```

```python
# i got the accuracy on both the trainng set and testing when using the gradient boosting classifoer
from sklearn.ensemble import GradientBoostingClassifier
grad = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

# Fit the classifier to the training data
grad.fit(x_train, y_train)

# Make predictions on the test data
y_pred25 = grad.predict(x_test)

print("Accuracy on training set before removing outliers: {:.3f}".format(grad.score(x_train, y_train)))
print("Accuracy on test set before removing outliers: {:.3f}".format(grad.score(x_test, y_test)))
```

```python
# this is just the confusion matrix of the gradient boosting before removing any outliers or 
#oversampling which shows why the class 2 is not predicted correctly that good due to that its 
#slightly imbalanced 
Confusion = confusion_matrix(y_test, y_pred25)
CM=ConfusionMatrixDisplay(Confusion)
CM.plot()
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

# Define a list of potential k values to try
k_values = range(1, 20)

# Initialize variables to store the best k and its corresponding accuracy
best_k = None
best_accuracy = 0.0

# it Iterates through each k value and find the one with the highest accuracy
for k in k_values:
    # i Created a K-Nearest Neighbors classifier with the current k value
    knn = KNeighborsClassifier(n_neighbors=k)
    # i Used cross-validation to estimate the model's accuracy
    accuracy = cross_val_score(knn, x_train, y_train, cv=5).mean()
    # Check if the current k gives a higher accuracy than the previous best
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_k = k

# Train the final K-Nearest Neighbors model using the best k
best_knn = KNeighborsClassifier(n_neighbors=best_k)
best_knn.fit(x_train, y_train)
# Make predictions on the test set
y_pred = best_knn.predict(x_test)

print("Best k:", best_k)
print("Accuracy on knn training set before removing the outliers: {:.3f}".format(best_knn.score(x_train, y_train)))
print("Accuracy on knn test set before removing the outliers: {:.3f}".format(best_knn.score(x_test, y_test)))
```

```python
#plot the confusion matrix of the best knn model that i got previously 
plot_confusion_matrix(best_knn,x_test,y_test,cmap='Purples')
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier

# Define a list of potential max_depth and min_samples_leaf values
max_depth_values = range(1, 20)
min_samples_leaf_values = range(1, 10)

# Initialize variables to store the best max_depth, min_samples_leaf, and accuracy
best_max_depth = None
best_min_samples_leaf = None
best_accuracy = 0.0

# Iterate through combinations of max_depth and min_samples_leaf and find the best ones
for max_depth in max_depth_values:
    for min_samples_leaf in min_samples_leaf_values:
        # Create a Decision Tree classifier with the current parameters
        dt = DecisionTreeClassifier(max_depth=max_depth, min_samples_leaf=min_samples_leaf)
        # Use cross-validation to estimate the model's accuracy
        accuracy = cross_val_score(dt, x_train, y_train, cv=5).mean()
         # Check if the current combination of parameters gives a higher accuracy
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_max_depth = max_depth
            best_min_samples_leaf = min_samples_leaf

# Train the final model with the best parameters
best_dt = DecisionTreeClassifier(max_depth=best_max_depth, min_samples_leaf=best_min_samples_leaf)
best_dt.fit(x_train, y_train)
# Make predictions on the test set
y_pred = best_dt.predict(x_test)

# Print the results
print("Best max_depth:", best_max_depth)
print("Best min_samples_leaf:", best_min_samples_leaf)
print("Accuracy on training set before removing the outliers: {:.3f}".format(best_dt.score(x_train, y_train)))
print("Accuracy on test set before removing the outliers: {:.3f}".format(best_dt.score(x_test, y_test)))
```

```python
#plot the confusion matrix of the best Decision tree model that i got previously 
plot_confusion_matrix(best_dt,x_test,y_test,cmap='Purples')
```

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, roc_auc_score, auc

# Assuming y_test is a one-hot encoded matrix or a list of labels
n_classes = len(np.unique(y_test))

# Define the models list and their corresponding labels
models = [log_clfs,rfs, knns, dts, grads]
model_labels = ['Logistic Regression','Random Forest', 'Knn', 'Decision Tree', 'Gradient Boosting']

# Create a figure for the ROC curve comparison plot
plt.figure(figsize=(8, 6))

# Plot ROC curve for each model
for model, label in zip(models, model_labels):
    probas_ = model.predict_proba(x_test)  # Calculate class probabilities (predicted probabilities) for each class
    
    # Binarize the true labels for each class
    y_test_bin = label_binarize(y_test, classes=np.unique(y_test))
    
    # Initialize variables to store the aggregate ROC curve data
    mean_fpr = np.linspace(0, 1, 100)
    mean_tpr = 0
    
    # Compute the ROC curve data for each class
    for i in range(n_classes):
        fpr, tpr, _ = roc_curve(y_test_bin[:, i], probas_[:, i], drop_intermediate=True)
        mean_tpr += np.interp(mean_fpr, fpr, tpr)
    
    # Average the ROC curve data across classes
    mean_tpr /= n_classes
    
    # Calculate the AUC (Area Under the Curve) for the averaged ROC curve
    mean_auc = auc(mean_fpr, mean_tpr)
    
    # Plot the ROC curve for the current model with its AUC value as part of the label
    plt.plot(mean_fpr, mean_tpr, label=f'{label} AUC: {mean_auc:.3f}')

# Customize the plot
plt.legend(loc=4)  # Add a legend in the lower-right corner
plt.xlabel('False Positive Rate')  # Label the x-axis as False Positive Rate
plt.ylabel('True Positive Rate')  # Label the y-axis as True Positive Rate
plt.title('Multiclass ROC Curves for Model Comparison')
plt.show()  # Display the ROC curve comparison plot
```
